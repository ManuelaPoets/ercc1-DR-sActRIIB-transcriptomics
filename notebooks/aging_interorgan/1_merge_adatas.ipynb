```python
import anndata
import numpy as np
```

```python
adata_mus = anndata.read_h5ad('/data/projects/manuela/aging_muscle/annotated-muscle-soupxed.h5ad')
adata_kid = anndata.read_h5ad('/data/projects/manuela/scRNA_aging-mouse/annotated-aging-soupxed.h5ad')
```

```python
# copy kidney celltypes into mixed_celltype column
adata_kid.obs['mixed_celltype'] = adata_kid.obs['celltype']

adata_kid.obs['tissue'] = 'kidney'
adata_mus.obs['tissue'] = 'muscle'
set(adata_mus.obs['mixed_celltype'])
```

```python
# Concatenate the AnnData objects
adata = anndata.AnnData.concatenate(adata_mus, adata_kid)
set(adata.obs['mixed_celltype'])
```

```python
conditions = [
    (adata.obs['sample'] == 'rgzj1'),  # Condition 1
    (adata.obs['sample'] == 'SG18'),  # Condition 2
    
    (adata.obs['sample'] == 'rgzj2'),
    (adata.obs['sample'] == 'SG20'),
    
    (adata.obs['sample'] == 'rgzj3'),
    (adata.obs['sample'] == 'SG24'),
    
    (adata.obs['sample'] == 'rgzj4'),
    (adata.obs['sample'] == 'SG26'),
    
    (adata.obs['sample'] == 'rgzj5'),
    (adata.obs['sample'] == 'SG28')]

# Corresponding values for the conditions
values = ['ctrl', 'ctrl', 'age', 'age', 'sAct', 'sAct', 'DR', 'DR', 'combi', 'combi']  # Values for Condition 1 and Condition 2

# Add more values as needed, ensuring they correspond to the conditions

# Default value if none of the conditions are met
default = 'other_condition'

# Add a new column "condition" based on the conditions and values
adata.obs['condition'] = np.select(conditions, values, default)
```

```python
adata_ctrl = adata[adata.obs['condition'] == 'ctrl']
adata_age = adata[adata.obs['condition'] == 'age']
adata_sAct = adata[adata.obs['condition'] == 'sAct']
adata_DR = adata[adata.obs['condition'] == 'DR']
adata_combi = adata[adata.obs['condition'] == 'combi']
```

```python
sub_adata = adata[adata.obs['mixed_celltype'].isin([ "PODO", "IMM", "EC-1(gEC)", "Myonuclei", "FAPs", "Macrophages"])].copy()
#sub_adata = adata[adata.obs['mixed_celltype'].isin([ "PODO", "IMM", "Myonuclei", "Endothelial cells", "Macrophages","Immune cells?"])].copy()


sub_sub_adata = adata[adata.obs['mixed_celltype'].isin([ "PODO"])].copy()

#set(adata.obs['mixed_celltype'])
#set(sub_adata.obs['mixed_celltype'])
set(sub_sub_adata.obs['mixed_celltype'])
```
