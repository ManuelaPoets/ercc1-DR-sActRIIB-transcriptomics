```r
library(SoupX)
library(Seurat)
library("MuDataSeurat")
library(ggplot2)
```

```r
packageVersion("SoupX")
```

## SG18

```r
# Calculate
sg18 = load10X('/data/projects/sonja/raw_data_sydney/mouse/SG18/outs')
sg18 = autoEstCont(sg18)
out_sg18 = adjustCounts(sg18)
```

```r
#Convert to seurat object and write to h5ad
seu_sg18 <- CreateSeuratObject(out_sg18, project = "sg18")

WriteH5AD(seu_sg18, "/data/projects/manuela/scRNA_aging-mouse/sg18_soupx.h5ad", assay="RNA")
```

```r
#Without SoupX
sc_sg18 <- Read10X('/data/projects/sonja/raw_data_sydney/mouse/SG18/outs/filtered_feature_bc_matrix')
seu_sg18 <- CreateSeuratObject(counts = sc_sg18)
WriteH5AD(seu_sg18, "/data/projects/manuela/scRNA_aging-mouse/sg18_unfil.h5ad", assay="RNA")
```
